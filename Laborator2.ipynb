```python
#Exercitiul 1.a

import numpy as np

# Starea initiala
urn = {
    "red": 3,
    "blue": 4,
    "black": 2
}

die_roll = np.random.randint(1, 7)

if die_roll in [2, 3, 5]:
    urn["black"] += 1
elif die_roll == 6:
    urn["red"] += 1
else:
    urn["blue"] += 1

balls = ["red"] * urn["red"] + ["blue"] * urn["blue"] + ["black"] * urn["black"]

drawn_ball = np.random.choice(balls)

print("Die roll: %d"%die_roll)
print("Updated urn: %s"%urn)
print("Drawn ball: %s"%drawn_ball)
```

```python
#Exercitiul 1.b, 1.c

import numpy as np

N=10000
red_count = 0

for _ in range(N):
  urn = { "red":3, "blue":4, "black":2}
  die_roll = np.random.randint(1,7)

  if die_roll in [2, 3, 5]:
    urn["black"] += 1
  elif die_roll == 6:
    urn["red"] += 1
  else:
    urn["blue"] += 1

  balls = ["red"] * urn["red"] + ["blue"] * urn["blue"] + ["black"] * urn["black"]

  drawn_ball = np.random.choice(balls)

  if drawn_ball == "red":
    red_count += 1;

p_red = red_count / N

print("Probabilitatea experimentala de a extrage o bila rosie: %.4f" %p_red)

p_teoretica =  1/2 * 3/10 + 1/6 * 4/10 + 1/3 * 3/10

print("Probabilitatea teoretica de a extrage o bila rosie: %.4f" %p_teoretica)
```

```python
#Exercitiul 2.1 2.2

import numpy as np
import matplotlib.pyplot as plt

lambdas = [1, 2, 5, 10]

N=1000

X1 = np.random.poisson(lam=1, size=N)
X2 = np.random.poisson(lam=2, size=N)
X3 = np.random.poisson(lam=5, size=N)
X4 = np.random.poisson(lam=10, size=N)


rand_lambda = np.random.choice(lambdas, size=N)

X5 = np.random.poisson(lam=rand_lambda)

```

```python
# Histograma pentru distributia X1

plt.hist(X1, bins=range(0, max(X1)+5), alpha=0.5, color='lightgreen', edgecolor='black')

plt.title("Histogram of Poisson Distributions")
plt.xlabel("Number of calls per hour")
plt.ylabel("Empirical density")
plt.show()
```
